```python
from pymongo import MongoClient
import os

client = MongoClient("mongodb+srv://ashlii:123567@cluster0.xukxbpk.mongodb.net/?appName=Cluster0")

DB_NAME = "langchain"
COLLECTION_NAME = "spider_snow_documents"
MONGODB_COLLECTION = client[DB_NAME][COLLECTION_NAME]
```

```python
import pandas as pd

documents = "Spider2-main/spider2-snow/resource/documents"

df = pd.read_json("Spider2-main/spider2-snow/spider2-snow.jsonl", lines=True)
docs = 0

for index, row in df.iterrows():
    # 如果external_knowledge字段为NaN，则跳过该行
    if pd.isna(row['external_knowledge']):
        continue

    # 构建知识文档的路径
    knowledge = f"{documents}/{row['external_knowledge']}"
    
    # 读取知识文档内容
    with open(knowledge, 'r', encoding='utf-8') as f:
        doc = {
            "_id" : row['instance_id'],
            "text": f.read(),
        }

    MONGODB_COLLECTION.insert_one(doc)
    print(f"Processed document: {knowledge}")
    docs += 1


print(f"Total documents processed: {docs}")

```
