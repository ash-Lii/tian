```python
from langchain_community.document_loaders import PyPDFLoader
import os

pdf_path = "data/postgis-3.6-en.pdf"
loader = PyPDFLoader(pdf_path)
documents = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500, chunk_overlap=100
)
docs = text_splitter.split_documents(documents)
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model=os.getenv("EMB_MODEL"))
```

```python
from langchain_postgres import PGVector

vector_store = PGVector(
    embeddings=embeddings,
    collection_name="postgis-3.5-en",
    connection=os.getenv("DB_URL").replace("...", "vector"),
)
```

```python
print(vector_store.collection_metadata)
```

```python
from tqdm import tqdm
n = len(docs)
pbar = tqdm(total=n)

for i, doc in enumerate(docs):  # 这里只添加前10个分块作为示例
    vector_store.add_documents(documents=[doc])
    pbar.update(1)
pbar.close()
```
